```python
import pandas as pd
import numpy as np  
# from scipy import find_peaks
import matplotlib.pyplot as plt
import cmath
import math as m
from scipy.signal import find_peaks, peak_widths, savgol_filter
from scipy.optimize import curve_fit, leastsq
from pylab import *
import glob
from datetime import *
```

```python
df=pd.read_csv(r"C:\Users\Anushka.PANDEY\Nextcloud\EPFL collab\sample1\short termination\VNA\Rawdata_4.88_GHz_0dB_200mK.txt", sep='\t', header=0)  
Frequency = df.iloc[:,0]          # 1st column
Simag = df.iloc[:, 1]       # 2nd column
Sreal = df.iloc[:, 2]       # 3rd column

# Combine into complex S21
S21_complex = np.array([complex(r, i) for r, i in zip(Sreal, Simag)])

# Compute magnitude in dB
S21_mag = np.abs(S21_complex)
S21_dB = 20 * np.log10(S21_mag)
S21_dB = S21_dB[1:]
f = Frequency[1:]
S21_complex = S21_complex[1:]
# Plot magnitude fit
plt.figure(figsize=(10, 6))
plt.plot(f, S21_mag[1:], label="Measured |S21|")
plt.xlabel("Frequency (Hz)")
plt.xlim(4.88e9-25000, 4.88e9+25000)
plt.ylabel("Magnitude |S21|")
plt.title("Magnitude S21")
plt.figtext(0.5, -0.05,'Rawdata_4.88_GHz_0dB_200mK', ha='center', fontsize=11, fontweight='bold')  
plt.legend()
plt.grid(True)
```

```python
filenames=["analyser_5.02034_GHz__pump_-51db(-5)_probe_-46dB(20)_200mK.txt", "analyser_5.02034_GHz__pump_-51db(-5)_probe_-61dB(5)_200mK.txt","analyser_5.97515_GHz__pump_-46db(0)_probe_-46dB(20)_200mK.txt"
,"analyser_5.101115_GHz__pump_-46db(0)_probe_-46dB(20)_200mK.txt"]

def read_measurement_file(filename):
    """Reads measurement data from CSV files"""
    folder = r"C:\Users\Anushka.PANDEY\Nextcloud\EPFL collab\sample1\short termination"
    freq = []
    S = []
    with open(folder+"\\"+filename, 'r') as f:
        data = f.readlines()
    
    # Read each line and extract frequency and power
    for line in data:
        fields = line.strip().split()
        if len(fields) != 2:
            continue
        try:
            freq.append(float(fields[0]))
            S.append(float(fields[1]))
        except ValueError:
            continue  # Skip lines that can't be parsed

    return np.array(freq), np.array(S)
for filename in filenames:
    freq, power = read_measurement_file(filename)
    freq = np.array(freq)
    power =np.array(power)
    # print(freq, power)
    plt.figure(figsize=(10, 6))
    plot(freq, power, label=f' ({filename.split("_")[1]} GHz)')
    plt.xlabel("Frequency (Hz)")
    plt.ylabel("Power (dBm)")
    # plt.xlim(-2e6, -1e6)
    plt.title("Mechanics")
    plt.figtext(0.5, -0.05, filename, ha='center', fontsize=11, fontweight='bold')
    plt.legend()
    plt.grid(True)
    show()

# # pumping at a higher frequency than resonance freq ( w + omega_m)
# filename_1 = "analyser_5.09972187GHz_pump-31db_RBW_1Hz_left_peak.txt"
# freq, power = read_measurement_file(filename_1)
# freq = np.array(freq)-5.09972187e9
# power = np.array(power)
# # print(freq, power)
# plt.figure(figsize=(10, 6))
# plot(freq, power)
# plt.xlabel("Frequency (Hz)")
# plt.ylabel("Power (dBm)")
# plt.title("Mechanics 5.1 Ghz 215mK{pumping at a higher frequency than resonance freq. ( w + omega_m)}")
# plt.figtext(0.5, -0.05, 'analyser_5.09972187GHz_pump-31db_RBW_1Hz_left_peak', ha='center', fontsize=11, fontweight='bold')
# plt.grid(True)
# show()

```
